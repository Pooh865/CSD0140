```python
from matplotlib import pyplot as plt
plt.style.use("seaborn-whitegrid")
import numpy as np
print("step 1")
```

```python
fig=plt.figure()
ax=plt.axes()
x=np.linspace(0,5,10)
# cosec=1/np.sin(x)
# sec=1/np.cos(x)
# cot=np.sin(x)/np.cos(x)
ax.plot(x,np.sin(x),label="SIN",color="orange",linestyle="dashdot")
ax.plot(x,np.cos(x),label="COS",color="yellowgreen",linestyle="dotted")
ax.plot(x,np.tan(x),label="TAN",color="red",linestyle="dashed")
ax.set_title("SIMPLE REPORT")
ax.set_xlabel("PERIOD")
ax.set_ylabel("AMPLITUDE")
ax.legend()
plt.show()
```
